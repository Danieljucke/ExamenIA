# Importation des d√©pendances

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import cluster
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression 
from sklearn.ensemble import StackingClassifier #combinaison des classifieurs
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
```

# Importation du dataset

```python
data = pd.read_csv(r"C:\Users\tjwri\Downloads\Credit-card-approval-prediction-classification-main\datasets\application_record.csv", index_col=0)
data
```

# nettoyage du dataset

```python
data['CODE_GENDER'].replace(['M','F'],[0,1], inplace=True)
data["FLAG_OWN_REALTY"].replace(['N','Y'],[0,1], inplace=True)
data["FLAG_OWN_CAR"].replace(['N','Y'],[0,1], inplace=True)
data["NAME_INCOME_TYPE"].replace(['Pensioner','Working','Commercial associate','State servant','Student'],[0,1,2,3,4], inplace=True)
data["NAME_EDUCATION_TYPE"].replace(['Secondary / secondary special','Higher education', 'Lower secondary','Incomplete higher','Academic degree'],[0,1,2,3,4], inplace=True)
data["NAME_FAMILY_STATUS"].replace(['Married','Civil marriage','Separated','Single / not married','Widow'],[0,1,2,3,4], inplace=True)
data["NAME_HOUSING_TYPE"].replace(['House / apartment','Rented apartment','Office apartment','Municipal apartment','Co-op apartment','With parents'],[0,1,2,3,4,5], inplace=True)
data =data.drop(['OCCUPATION_TYPE'],axis= 1)
```

# Cr√©ation du model pour avoir les sorties

```python
model_kmeans=cluster.KMeans(n_clusters=2)
```

```python
Results=model_kmeans.fit_predict(data)
Results
```

```python
data["Resultats"]=Results
```

```python
# data.to_csv('DatasetTreated')
data
```

```python
Y=data["Resultats"]
X=data.drop(["Resultats"], axis=1)
X
```

# Split The Dataset

```python
X_train, X_test, Y_train, Y_test= train_test_split(X,Y,test_size=0.20,random_state=0)
```

# Logistic Regression

```python
model_LR= LogisticRegression()
```

```python
model_LR.fit(X_train,Y_train)
```

```python
y_train_pred= model_LR.predict(X_test)
```

```python
model_LR.score(X_test,Y_test)
```

# Decision Tree

```python
Decision_tree = DecisionTreeClassifier(criterion='entropy')
```

```python
Decision_tree.fit(X_train,Y_train)
```

```python
Decision_tree.predict(X_test)
```

```python
Decision_tree.score(X_test,Y_test)
```

# Stacked Model

```python
listes_estimateurs = [
    ('dt',Decision_tree),
    ('LR',model_LR) 
]
```

```python
stack_model = StackingClassifier(
    estimators=listes_estimateurs, final_estimator=RandomForestClassifier()
)
```

```python
stack_model.fit(X_train, Y_train)
```

```python
stack_model.predict(X_test)
```

```python
stack_model.score(X_test,Y_test)
```
